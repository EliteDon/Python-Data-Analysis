```python
import pandas as pd
import numpy as np
```

```python
# students = pd.DataFrame({"":{},"":{},"":{},"":{}})也可以这样建
name = pd.Series({"001":"小陈","002":"小王","003":"小李","004":"小马","005":"小二","006":"小八"})
test1 = pd.Series({"001":78,"002":55,"003":65,"004":54,"005":76,"006":43})
test2 = pd.Series({"001":75,"002":52,"003":61,"004":55,"005":72,"006":45})
test3 = pd.Series({"001":71,"002":53,"003":64,"004":56,"005":71,"006":47})
students = pd.DataFrame({"姓名":name,"考试1":test1,"考试2":test2,"考试3":test3})
students
```

```python
average = students.loc[:,"考试1":"考试3"].mean(axis = 1)
student_average = pd.DataFrame({"姓名":name,"考试成绩平均分":average})
student_average
```

```python
def score_to_level(score):
    if score > 90:
        return "A"
    elif (score >80 and score <= 90):
        return "B"
    else:
        return "C"

students.loc[:,"考试1":"考试3"].map(score_to_level)
```

```python
students["考试1"] = students["考试1"].astype('int')
students["考试2"] = students["考试2"].astype('int')
students["考试3"] = students["考试3"].astype('int')

students.describe()
```
