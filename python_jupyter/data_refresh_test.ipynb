## 清洗数据

```python
cleaned_data = oringinal_data.copy()
```

```python
cleaned_data.head()
```

```python
# 数据类型转换
# 转换成日期
# cleaned_data[""] = pd.to_datatime(cleaned_data[""])
```

```python
# 转换成字符串
cleaned_data["Age"] = cleaned_data["Age"].astype(str)
cleaned_data["Age"]
```

```python
# 去小数点slice是留头不留尾
cleaned_data["Age"].str.slice(0,-2)
```

```python
# 删除dataframe的缺失值
# cleaned_data.dropna(subset=[""],inplace=True)
```

```python
# 替换
cleaned_data["Player"] = cleaned_data["Player"].replace({"LeBron James":"LBJ"})
```

```python
cleaned_data.head()
```

```python
len(cleaned_data[cleaned_data["Player"] == "LBJ"])
```

```python
cleaned_data = cleaned_data[cleaned_data["Height"].notnull()]
```

```python
cleaned_data.info()
```

```python
cleaned_data = cleaned_data[cleaned_data["Weight"].notnull()]
cleaned_data = cleaned_data[cleaned_data["Age"].notnull()]
cleaned_data = cleaned_data[cleaned_data["Player"].notnull()]
cleaned_data = cleaned_data[cleaned_data["Number"].notnull()]
```

```python
cleaned_data.info()
```
